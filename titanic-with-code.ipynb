```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
titanic = pd.read_csv("data/titanic.csv")
titanic
```

```python
titanic.plot.hist(column="Age", grid=False, edgecolor='black', xlabel="Age");
```

```python
titanic['Sex'].value_counts().plot(kind='pie', figsize=(8, 8), title='Distribution of Passenger Sex', autopct='%1.1f%%');
```

```python
titanic.plot.hist(column="Fare", bins=range(0,600,50), grid=False, edgecolor='black', xlabel="Fare");
```
